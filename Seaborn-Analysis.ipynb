```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
s=pd.read_csv("Hotel.csv")
n=s.head(50)
r=s.tail(50)
fig=plt.figure(figsize=(25,5))
sns.countplot(data=n,x="Hotel_name",hue="Rating",palette="viridis",width=1.5)
plt.tight_layout()
sns.pairplot(data=n,hue="Reviews",aspect=2,dropna=True,palette="rainbow")
```

```python
fig=plt.figure(figsize=(20,5))
sns.swarmplot(data=r,x="Hotel_name",y="Rating")
```
